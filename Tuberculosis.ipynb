### Estimated prevalence of TB (all forms) per 100 000 population,

```python
# get data only for the year 2016
df_1990 = df.query("Year == 1990" )
```

```python
df_1990.head()
```

```python
df_1990.tail()
```

```python
# create a choropleth using Plotly
fig = px.choropleth(
    df_1990,
    locations = 'ISO 3-character country/territory code',
    color = 'Estimated prevalence of TB (all forms) per 100 000 population',
    hover_name = 'Country or territory name',
    color_continuous_scale = px.colors.sequential.Jet,
    width = 1000,
    height = 1000
)

fig.update_layout (
    title_text = 'Estimated prevalence of TB (all forms) per 100 000 population across the world in 1990',
    width = 2000,
    height = 1000,
)

fig.show()
```

```python
# add animation to our plot
fig = px.choropleth(
    df,
    locations = 'ISO 3-character country/territory code',
    color = 'Estimated prevalence of TB (all forms) per 100 000 population',
    hover_name = 'Country or territory name',
    animation_frame = 'Year',
    color_continuous_scale = px.colors.sequential.Jet,
)

fig.update_layout (
    title_text = 'Estimated prevalence of TB (all forms) per 100 000 population across the years',
    width = 2000,
    height = 1000,
    geo = dict(projection = {'type' : 'mollweide'})
)

fig.show()
```
